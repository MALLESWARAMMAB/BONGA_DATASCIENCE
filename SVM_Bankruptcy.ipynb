```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
import numpy as np

# Split data into training and testing sets initially without stratification
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# Check if stratification is possible based on class distribution in training set
unique_classes_train, class_counts_train = np.unique(Y_train, return_counts=True)
stratify_param = Y if np.all(class_counts_train > 1) else None  # Check if all classes have at least 2 samples

# Re-split with stratification if possible, otherwise without
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=stratify_param, random_state=0)

# Apply SMOTE only if Y_train has more than one class
if len(unique_classes_train) > 1:
    smote = SMOTE(random_state=42)
    X_resampled, Y_resampled = smote.fit_resample(X_train, Y_train)
    # Optional: assign X_resampled, Y_resampled back to X_train, Y_train if desired
    X_train, Y_train = X_resampled, Y_resampled
```
